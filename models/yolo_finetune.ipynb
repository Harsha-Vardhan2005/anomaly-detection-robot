```python
from ultralytics import YOLO
import cv2
import os

# Load the trained model
model = YOLO('/content/runs/detect/suspicious_detection_exp/weights/best.pt')

# Predict on test images
results = model.predict(
    source='/content/drive/My Drive/suspicious detection.v1i.yolov8/test/images',
    save=False,  # Don’t save automatically, we’ll handle it
    conf=0.5     # Confidence threshold
)

# Output directory (create if it doesn’t exist)
output_dir = '/content/runs/detect/custom_alerts'
os.makedirs(output_dir, exist_ok=True)

# Process each result and add alert text for weapons
for result in results:
    # Get the original image
    img = result.orig_img.copy()  # Copy of the image
    weapon_detected = False

    # Check for weapon detections
    for box in result.boxes:
        class_id = int(box.cls[0])
        label = model.names[class_id]
        conf = box.conf[0]
        if label == 'weapon' and conf >= 0.5:
            weapon_detected = True
            # Get bounding box coordinates
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            # Draw the bounding box
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 2)  # Red box
            # Add label and confidence above the box
            cv2.putText(img, f"{label} {conf:.2f}", (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

    # If weapon detected, add alert text
    if weapon_detected:
        cv2.putText(img, "Alert! Weapon Detected", (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)  # Red text at top-left

    # Save the modified image
    save_path = os.path.join(output_dir, os.path.basename(result.path))
    cv2.imwrite(save_path, img)

print(f"Results saved to {output_dir}")

# Zip and download
!zip -r predictions_with_alerts.zip {output_dir}
from google.colab import files
files.download('predictions_with_alerts.zip')
```

```python
from ultralytics import YOLO
import cv2
import os

# Load the trained model
model = YOLO('/content/runs/detect/suspicious_detection_exp/weights/best.pt')

# Predict on test images
results = model.predict(
    source='/content/drive/My Drive/suspicious detection.v1i.yolov8/test/images',
    save=False,  # We’ll handle saving manually
    conf=0.5     # Confidence threshold
)

# Output directory
output_dir = '/content/runs/detect/custom_alerts'
os.makedirs(output_dir, exist_ok=True)

# Process each result
for result in results:
    # Get the original image
    img = result.orig_img.copy()
    weapon_detected = False

    # Annotate all detected objects
    for box in result.boxes:
        class_id = int(box.cls[0])
        label = model.names[class_id]
        conf = box.conf[0]
        x1, y1, x2, y2 = map(int, box.xyxy[0])

        # Draw bounding box and label for all classes
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green box for all
        cv2.putText(img, f"{label} {conf:.2f}", (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # Check for weapon specifically
        if label == 'weapon' and conf >= 0.5:
            weapon_detected = True
            # Redraw weapon box in red for emphasis
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 2)
            cv2.putText(img, f"{label} {conf:.2f}", (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

    # Add alert text if weapon detected
    if weapon_detected:
        cv2.putText(img, "Alert! Weapon Detected", (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)  # Red alert text

    # Save the modified image
    save_path = os.path.join(output_dir, os.path.basename(result.path))
    cv2.imwrite(save_path, img)

print(f"Results saved to {output_dir}")

# Zip and download
!zip -r predictions_with_alerts.zip {output_dir}
from google.colab import files
files.download('predictions_with_alerts.zip')
```

```python
from ultralytics import YOLO
import cv2
import os
from gtts import gTTS
from IPython.display import Audio  # To play audio in Colab

# Install mpg123 (run this once if not done)
!apt-get install mpg123

# Load the model
model = YOLO('/content/runs/detect/suspicious_detection_exp/weights/best.pt')

# Predict on test images
results = model.predict(
    source='/content/drive/My Drive/suspicious detection.v1i.yolov8/test/images',
    save=False,
    conf=0.5
)

# Output directory
output_dir = '/content/runs/detect/custom_alerts'
os.makedirs(output_dir, exist_ok=True)

# Process each result
for i, result in enumerate(results):
    img = result.orig_img.copy()
    weapon_detected = False
    high_threat = False

    # Annotate all detections
    for box in result.boxes:
        class_id = int(box.cls[0])
        label = model.names[class_id]
        conf = box.conf[0]
        x1, y1, x2, y2 = map(int, box.xyxy[0])

        # Default green for all
        color = (0, 255, 0)
        if label == 'weapon':
            weapon_detected = True
            color = (0, 0, 255)  # Red for weapon
            for other_box in result.boxes:
                if model.names[int(other_box.cls[0])] == 'victim':
                    ox1, oy1, ox2, oy2 = map(int, other_box.xyxy[0])
                    if abs(x1 - ox1) < 50 and abs(y1 - oy1) < 50:
                        high_threat = True

        cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)
        cv2.putText(img, f"{label} {conf:.2f}", (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    # Add severity-based alert and audio
    if weapon_detected:
        alert_text = "Caution! Weapon Detected" if not high_threat else "High Alert! Threat Detected"
        alert_color = (0, 255, 255) if not high_threat else (0, 0, 255)
        cv2.putText(img, alert_text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, alert_color, 2)

        # Generate unique audio file for this image
        audio_file = f"alert_{i}.mp3"
        tts = gTTS(alert_text)
        tts.save(audio_file)

        # Play audio in Colab
        display(Audio(audio_file, autoplay=True))  # Plays in notebook

        # Move audio to output directory
        os.rename(audio_file, os.path.join(output_dir, audio_file))

    # Save image
    save_path = os.path.join(output_dir, os.path.basename(result.path))
    cv2.imwrite(save_path, img)

print(f"Results saved to {output_dir}")
!zip -r predictions_with_alerts.zip {output_dir}
from google.colab import files
files.download('predictions_with_alerts.zip')
```
